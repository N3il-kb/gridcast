```python
import pandas as pd
import plotly.express as px


data = {
        'City': ['London', 'Paris', 'New York'],
        'Latitude': [51.5, 48.8, 40.7],
        'Longitude': [-0.1, 2.3, -74.0],
        'Population': [8.9, 2.1, 8.4],
        'Continent': ['Europe', 'Europe', 'North America']
    }
df = pd.DataFrame(data)
```

```python
df = pd.read_csv('datacenter_scores.csv')
df
```

```python
fig = px.scatter_geo(
        df,
        lat="lat",  # Column containing latitude values
        lon="lon", # Column containing longitude values
        size="hybrid", # Column to determine marker size
        hover_name="region", # Column to display on hover
        projection="natural earth", # Map projection type
    )
fig.show()
```
